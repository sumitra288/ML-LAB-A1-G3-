# [Objective 03](#pandas-visualizations) Describe and discriminate between basic data types

## Overview

One of the cornerstones of Exploratory Data Analysis (EDA) is being able to identify variable types.  We will need different statistical methods to display and describe each of these different types of data.

```
df.isnull().sum()
```

```

df.isnull()
```

```

df.columns
```

```
#the species in caps are treated differently so changing all the species in small
df['Species'] = df['Species'].str.lower()
df['Species'].value_counts()
```

```

[features for features in df.columns if df[features].isnull().sum()>0]
```

```

sns.heatmap(df.isnull(), yticklabels=10, cbar=True, cmap = 'viridis')
```

```

df['Sepal.Length']=df['Sepal.Length'].fillna(df['Sepal.Length'].mean())
df['Sepal.Length'].isnull().sum()
```

```
df['Sepal.Width']=df['Sepal.Width'].fillna(df['Sepal.Width'].mean())
df['Sepal.Width'].isnull().sum()
```

```
df.isnull().sum()
```

```

sns.heatmap(df.isnull(), yticklabels=10, cmap='viridis', cbar=True)
#no null value is present
```

```
sns.pairplot(df.drop(columns='Unnamed: 0', errors='ignore'),
             hue='Species', corner=True, markers=["o", "s", "D"])
```

```

sns.heatmap(df.drop("Species", axis=1).corr(), annot=True, cmap="coolwarm")
```

```

sns.lmplot(x="Petal.Length", y="Petal.Width", hue="Species", data=df)
```

```

df.groupby("Species").mean()
```
